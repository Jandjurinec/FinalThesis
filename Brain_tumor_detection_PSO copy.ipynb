```python
from PSO import *
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

if tf.config.list_physical_devices('GPU'):
    print("CUDA is available!")
else:
    print("CUDA is not available.")
```

```python
bs, cc, ccp = PSO(swarm_size=20, max_iter=30,
            c1=2, c2=2.05, w=0.5,
            is_global_neigh=False)
```

```python
iterations = np.arange(len(cc))

plt.figure(figsize=(8, 6))
plt.plot(iterations, cc, marker='o', linestyle='-', color='b', label='Objective Value')
plt.grid(visible=True)
plt.xlabel('Iterations')
plt.ylabel('F1')
plt.show()
```

```python
iterations = np.arange(len(ccp))

plt.figure(figsize=(8, 6))
plt.plot(iterations, ccp, marker='o', linestyle='-', color='b', label='Objective Value')
plt.grid(visible=True)
plt.xlabel('Iterations')
plt.ylabel('F1')
plt.show()
```
