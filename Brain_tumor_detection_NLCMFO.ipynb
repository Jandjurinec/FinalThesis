```python
from NLCMFO import *
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

if tf.config.list_physical_devices('GPU'):
    print("CUDA is available!")
else:
    print("CUDA is not available.")
```

```python
bf, cc, ccm = NLCMFO(alpha=1.8, r=4, num_of_moths=30, MaxIter=10, max_num_flames=15) # ~175 min
```

```python
iterations = np.arange(len(cc))

plt.figure(figsize=(8, 6))
plt.plot(iterations, cc, marker='o', linestyle='-', color='b', label='Objective Value')
plt.grid(visible=True)
plt.xlabel('Iterations')
plt.ylabel('F1')
plt.show()
```
